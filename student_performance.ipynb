```python
df.drop("student_id", axis=1, inplace=True)
```

```python
def plots (var,num):
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde= True )

plots('weekly_self_study_hours', 1)
plots('attendance_percentage', 2)
plots('class_participation', 3)
plots('total_score', 4)

plt.tight_layout()
```
